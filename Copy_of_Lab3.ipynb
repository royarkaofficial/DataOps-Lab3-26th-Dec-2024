GroupBy

```python
import pandas as pd

data = {
     'dept' : ['sales' , 'sales' ,'marketing','marketing','HR','HR'],
     'emp' : ['Shravya','Vikas','Srinivasa','Sonu','Ravi','Rahul'],
     'salary' : [50000,60000,70000,75000,65000,55000]
 }
df = pd.DataFrame(data)
df
```

Grouping them by Department and Calculate the mean salary for each department

```python
avg_sal = df.groupby('dept')['salary'].mean()
avg_sal
```

Total salary for each department

```python
total_sal = df.groupby('dept')['salary'].sum()
total_sal
```

Count the number of employees in every department

```python
count = df.groupby('dept')['emp'].count()
count
```

Summary of all the GroupBy Clauses

```python
summary = df.groupby('dept').agg({'salary':['mean','sum'],'emp':'count'})
summary
```

Adding the custom function

```python
salary_range = df.groupby('dept')['salary'].apply(lambda x: x.max() - x.min())
salary_range
```

**Analyzing a Sales Dataset**

Problem : Given a dataset of monthly transactions with columns as Product , Date , Region , Sales & Quantity. Analyze the data to answer :

```python
import pandas as pd
data = pd.read_csv('/content/sample_data/synthetic_sales_data.csv')
data
```

1. Which product had the highest sales in each region

```python
highest_salary_product = data.groupby(['Region','Product name'])['Sales'].sum().reset_index().sort_values('Sales',ascending=False).drop_duplicates('Region')
highest_salary_product
```

```python
high = data.groupby(['Region','Product name'])['Sales'].sum()
high
```

```python
high = high.reset_index()
high
```

```python
high = high.sort_values('Sales',ascending = False )
high
```

```python
high = high.drop_duplicates('Region')
high
```

2. What is the monthly sales trend for each product

```python
data['Yearmont'] = pd.to_datetime(data['Date'],format='%Y-%m-%d')
data
```

```python
ms = data.groupby([data["Yearmont"].dt.to_period("M"),'Product name'])['Sales'].sum()
ms
```

```python
ms = ms.unstack()
ms
```

```python
ms = ms.fillna(0)
ms
```
